## 4. How good does behaving completely random do?

```python
state=env.reset()
counter=0
g=0
reward=None
```

```python
while reward !=20:
    state,reward,terminated,truncated,info=env.step(env.action_space.sample())
    counter+=1
    g+=reward
```

```python
print("Solved in {} steps with a total reward of {}".format(counter,g))
```
