```python
print("PRANAL BORSE - 221A026")
import numpy as np
import random
```

```python
print("PRANAL BORSE - 221A026")

class GridWorld:
  def __init__(self,size =4 ,start=(0,0),goal=(3,3)):
    self.size=size
    self.start=start
    self.goal=goal
    self.state=start
    self.actions=[(0,1),(1,0),(0,-1),(-1,0)]  #Right,down,left ,up

  def reset(self):
    self.state=self.start
    return self.state

  def step (self,action):
    next_state = (self.state[0]+self.actions[action][0],
                  self.state[1]+self.actions [action][1])

    if 0<= next_state[0]< self.size and 0 <= next_state[1] <self.size:
      self.state = next_state
      reward =1 if self.state ==self.goal else -0.1
      done = self.state ==self.goal
      return self.state,reward,done
    else:
      # If the move is invalid, the state doesn't change, apply a penalty, and it's not done.
      reward = -1 # Penalty for hitting a wall
      done = False
      return self.state, reward, done
```

#2. Monte carlo control with exploring starts

```python
print("PRANAL BORSE - 221A026")

def monte_carlo_control (env,episodes=5000,gamma=0.9,epilson=0.1):
 Q={((i,j), a): 0 for i in range (env.size) for j in range (env.size)
 for a in range (4)}
 returns = {((i,j), a): [] for i in range (env.size) for j in range (env.size)
 for a in range (4)}

 for _ in range(episodes):
  state = env.reset()
  episode = []

  while True:
    # Choose action using epsilon-greedy policy derived from Q
    if random.random() < epilson:
        action = random.choice(range(4))
    else:
        action = max(range(4), key=lambda a_val: Q[(state, a_val)])

    next_state, reward, done = env.step(action)
    episode.append((state, action, reward))
    state = next_state
    if done:
      break

  G = 0
  # Iterate through the episode in reverse to calculate returns
  for state_t, action_t, reward_t in reversed(episode):
    G = gamma * G + reward_t
    # Check if (state, action) pair has been visited for the first time in this episode
    first_occurrence = True
    for s, a, _ in episode[:episode.index((state_t, action_t, reward_t))]: # corrected slicing
        if s == state_t and a == action_t:
            first_occurrence = False
            break
    if first_occurrence:
        returns[(state_t, action_t)].append(G)
        Q[(state_t, action_t)] = np.mean(returns[(state_t, action_t)])

 return Q
```

#3.SARSA (State-Action-Reward-State)

```python
print("PRANAL BORSE - 221A026")

#Temporal difference learning algorithm (SARSA)
def sarsa (env,episodes=5000,gamma=0.9,alpha=0.1,epilson =0.1):
  Q={((i,j), a): 0 for i in range (env.size) for j in range (env.size)
  for a in range (4)}

  for _ in range(episodes):
    state = env.reset()
    action = random.choice(range(4)) if random.random() < epilson else max(range(4), key=lambda a_val: Q[(state, a_val)])
    while True:
      next_state, reward, done = env.step(action)
      next_action = random.choice(range(4)) if random.random() < epilson else max(range(4) , key=lambda a: Q[(next_state, a)])

      Q[(state, action)] += alpha * (reward + gamma * Q[(next_state, next_action)] - Q[(state, action)])
      state,action = next_state,next_action
      if done:
        break
  return Q
#
```

Q- Learning is very similar to SARSA but uses max feature Q -value instead of followig the policy

#4.Q Learning Algotrithm

```python
print("PRANAL BORSE - 221A026")

#Q-Learning algorithm
def q_learning(env,episodes=5000,gamma=0.9,alpha=0.1,epilson=0.1):
  Q={((i,j), a): 0 for i in range (env.size) for j in range (env.size)
  for a in range (4)}

  for _ in range(episodes):
    state = env.reset()

    while True:
      action = random.choice(range(4)) if random.random() < epilson else max(range(4), key=lambda a: Q[(state, a)])
      next_state, reward, done = env.step(action)
      # Q-learning update: uses max over next actions
      Q[(state, action)] += alpha * (reward + gamma * max(Q[(next_state, a)] for a in range(4)) - Q[(state, action)])
      state = next_state
      if done:
        break
  return Q
```

#5.Running the training

```python
print("PRANAL BORSE - 221A026")

#Initialize Environment
env = GridWorld()

#Train with Monte Carlo
Q_mc = monte_carlo_control(env)

#Train with SARSA
Q_sarsa = sarsa(env)

#Train with Q_Learning
Q_qlearning = q_learning(env)
```
